Please create a simple adder using Tkinter. 
- The GUI should look like the image below. Your widgets could be at arbitrary width and height. But their relatively position should be the same as below. <br>
- The user should be able to input two numbers via two text entries. You can safely assume users will input valid integer number
- When the user click the button with equal sign, the result should be shown at the right-most label. 

```python
import tkinter as tk

# https://python-course.eu/tkinter/events-and-binds-in-tkinter.php

def on_calc(event):
    #CALCULATE THE ANSWER
    num1 = int(ent1.get())
    num2 = int(ent2.get())
    ans =  num1 + num2
    lab2["text"] = str(ans)

window = tk.Tk()
#window.title("Simple Adder")
    
ent1 = tk.Entry(window)
ent1.pack(side=tk.LEFT)
#ent1.grid(row=0, column=0, padx=10, pady=10)

label = tk.Label(window, text="+")
label.pack(side=tk.LEFT)
#label.grid(row=0, column=1)


ent2 = tk.Entry(window)
ent2.pack(side=tk.LEFT)
#ent2.grid(row=0, column=2, padx=10, pady=10)


bnt1 = tk.Button(window, text="=")
bnt1.pack(side=tk.LEFT)
bnt1.bind("<Button-1>", on_calc)  # Bind the function to the button's click event
#bnt1.grid(row=0, column=3, padx=10, pady=10)


lab2 = tk.Label(window)
lab2.pack(side=tk.LEFT) # .pack method adds the label to the window.
#lab2.grid(row=0, column=4, padx=10, pady=10)


window.mainloop()
```

Use Tkinter to create the GUI below. The UI contains one label and two buttons. When click "add 1", the number should increase by 1. When click the "minus 1" button, the number on the label should decrease by 1. <br>

The dimension of all three widgets should be width = 10, height = 5

```python
import tkinter as tk

ans = 0

def on_add(event):
    global ans
    ans += 1
    lab.config(text=str(ans))
    
def on_minus(event):
    global ans
    ans -= 1
    lab.config(text=str(ans))
    
window = tk.Tk()
window.title("Number Counter")
window.configure(bg="white")

        
bnt1 = tk.Button(
    text = 'minus 1',
    width = 10,
    height = 5,
    bg="#F0FFFF"
)
bnt1.pack(side=tk.LEFT)

bnt2 = tk.Button(
    text = 'add 1',
    width = 10,
    height = 5,
    bg="#F0FFFF"
)
bnt2.pack(side=tk.LEFT)

lab = tk.Label(
    text = '0',
    width = 10,
    height = 5
)

lab.pack(side=tk.LEFT)

# INSERT YOUR CODE BELOW
# Bind the on_button_click to the button
# ~ 2 lines
bnt1.bind("<Button-1>", on_minus)
bnt2.bind("<Button-1>", on_add)

window.mainloop()
```

Task 3: GO board

<!-- BEGIN QUESTION -->

Let's create a basic 9 x 9 Go board for beginner. 



The UI above use the grid layout with 81 buttons. When the user clicks the button, the button should sequentially change its content into either black or white piece. 

Some hints below:
- you can draw the Go pieces with unicode \u26AB and \u26AA
- you can access the widget the user is interacting with via **event.widget** in the callback function.

```python
import tkinter as tk

turn = 0
win = tk.Tk()

def on_btn_click(event):
    global turn

    if event.widget["text"] != '':
        return # do nothing if the user clicked on a filled position

    # Decide wheather to update the text to black or white piece
    # ~ 5 lines
    if turn % 2 == 0:
        event.widget["text"] = '\u26AA'  # White piece
    else:
        event.widget["text"] = '\u26AB'  # Black piece

    turn += 1
    
n = 9
for i in range(n):
    for j in range(n):
        # Create buttons nad use grid layout
        btn = tk.Button(width = '5',height = '2',text = '')
        # USER GRID LAYOUT AND BIND FUNCTIONS BELOW
        # ~ 2 lines
        btn.grid(row=i, column=j)
        btn.bind("<Button-1>", on_btn_click)

win.mainloop()

```
